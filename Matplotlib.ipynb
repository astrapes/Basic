```python
from matplotlib import rc    #책에 있는 한글 입력 방법
rc('font',family = 'Malgun Gothic')

import matplotlib as mpl     #ppt에 있는 한글 입력 방법
mpl.rcParams['axes.unicode_minus'] = False    #minus 표시
mpl.rc('font',family = 'Malgun Gothic')

font1 = {'size':18,'color':'green'}

xdata = ['안지영','홍지수','황예린']
ydata = [90,85,88]

plt.plot(xdata,ydata)

plt.title('세 명 학생의 국어 성적', fontdict=font1)

plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```

```python
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]

plt.plot(xdata,ydata1, label='국어')
plt.plot(xdata,ydata2, label='영어')
plt.legend(loc='upper center')

plt.title('세 명 학생의 국어,영어 성적')

plt.show()
```

```python
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]


plt.plot(xdata,ydata1, label='국어',color='red',linestyle='-',marker='o')
plt.plot(xdata,ydata2, label='영어',color='#00ffff',linestyle='--',marker='x')
plt.plot(xdata,ydata3, label='수학',color='magenta',linestyle='-.',marker='s')
plt.plot(xdata,ydata4, label='사회',color='#444444',linestyle=':',marker='d')

plt.legend(loc='best')

plt.title('세 명 학생의 네 과목 성적')

plt.show()
```

```python
plt.rcParams["font.size"] = 15

plt.plot(xdata,ydata1,'ro-',label='국어',linewidth =3,markersize =10)
plt.plot(xdata,ydata2, 'cx--', label='영어',linewidth =3,markersize =10)
plt.plot(xdata,ydata3, 'ms-.',label='수학',linewidth =3,markersize =10)
plt.plot(xdata,ydata4, 'kd:',label='사회',linewidth =3,markersize =10)

plt.legend(loc='best')

plt.title('세 명 학생의 네 과목 성적')

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,10)
plt.rcParams["font.size"] = 30
xdata = list(range(30))
ydata=[]

for x in xdata :
    y= 2*x
    ydata.append(y)
    
plt.title('y=2x')
plt.plot(xdata,ydata,'md-',label='y=2x')
plt.xlim(0,35)
plt.ylim(0,70)
plt.xticks(list(range(-2,32,2)))
plt.yticks(list(range(-2,60,5)))
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams["font.size"] = 15

a = np.arange(0,2*np.pi,0.2)
#a = linspace(0,(2*np.pi),500)
s = np.sin(a)
c = np.cos(a)

plt.title('sin,cos')
plt.plot(s,label='sin')
plt.plot(c,label='cos')
plt.xticks(np.arange(0,2*np.pi,np.pi/2))
plt.legend(loc='best')


plt.grid(True)

plt.show()
```

```python
xs = np.linspace(-3,3,601)
ys = 1 / (1 + np.exp(-xs))

plt.plot(xs,ys)
plt.ylim(-0.1,1.1)
plt.title('시그모이드 그래프')
plt.axhline(y=0,color='k',linewidth=0.5)
plt.axhline(y=1,color='k',linewidth=0.5)
plt.axhline(x=0,color='k',linewidth=0.5)
plt.show()
```
